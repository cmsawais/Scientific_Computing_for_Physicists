Vector sum of vec= [1.0, 1.0E16, -1.0E-16, -0.5] using three different methods. (for loop, sum functions and Kahan Summation Algorithm)

```python
## Mthod 1: Using For Loop ###
a = [1.0, 1.0e16, -1.0e16, -0.5]
sum_for = 0.0
for v in a:
    sum_for += v
print ("Sum using for loop:", sum_for) 
```

```python
#### Mthod 2 Using Sum function  ###############
import numpy as np
a = np.array([1.0, 1.0e16, -1.0e16, -0.5])
sum_numpy = np.sum(a)
print ("Sum using numpy.sum function is", sum_numpy)
```

```python
# Method 3: Kahan Summation Algorithm
a = [1.0, 1.0e16, -1.0e16, -0.5]

def kahan_sum(input_list):
    total = 0.0
    compensation = 0.0  # to keep track of low-order bits

    for value in input_list:
        y = value - compensation
        t = total + y
        compensation = (t - total) - y
        total = t

    return total  # âœ… this is critical!

sum_kahan = kahan_sum(a)
print("Sum using Kahan summation:", sum_kahan)
```

 Rewrite the daxpy algorithms to fill the vectors x and y with a gaussian random variable with mean zero and standard deviation 1. 

```python
import numpy as np

# Parameters
N = 1000          # Vector length
a = 3.0            # Scalar

# Generate Gaussian vectors x and y
x = np.random.normal(loc=0.0, scale=1.0, size=N)
y = np.random.normal(loc=0.0, scale=1.0, size=N)

# Perform DAXPY: d = a * x + y
d = a * x + y
print (d)

# Check and print result
print("First  value of x:", x[:1])
print("First value of y:", y[:1])
print("First value of d = ax + y:", d[:1])

# Optional check
if np.allclose(d, a * x + y):
    print(" DAXPY computation successful.")
```
